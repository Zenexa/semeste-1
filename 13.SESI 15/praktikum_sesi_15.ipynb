# **visualisasi data dengan matplotlib.pyplot**
# **mengroupkan data dengan groupby()**
# **mean() untuk menghitung nilai rata-rata**
# **mengubah data ke interger dengan astype(int)**




````python
data_group = fl_1.groupby('tahun')[['harga_juta','tahun_1','tahun_2']].mean().astype(int)

data_group.plot(kind = 'bar') #visualisasi data dalam bentuk bar
mt.xlabel('tahun') # memberi label x
mt.ylabel('Harga') # memberi label y
mt.title('perubahan harga dari 2015 > 2020,perubahan data 1 tahun kede pan dan 2 tahun kedepan') # memberi judul
mt.show() # untuk menampilkan visualisasi
````
